```python
import numpy as np
import pandas as pd
```

```python
import json

file = open('MitreData.json')

dataset = []

for line in file:
    dataset.append(eval(line))
    
df = pd.DataFrame(dataset)
```

```python
df['combined'] = df.iloc[:,df.columns != 'id'].apply(lambda x: ". ".join(x.astype(str)), axis=1)
```

```python
from nltk import tokenize

for i in range(0, len(df['combined'])):
    df['combined'][i] = tokenize.sent_tokenize(df['combined'][i])
```

```python
mappings = []
```

```python
for i in range(0, len(df['combined'])):
    for sentence in df['combined'][i]:
        senetence = sentence.strip()
        if sentence not in ['nan.', 'nan']:
            mappings.append({"description" : sentence, "id" : df['id'][i]})
```

```python
newDf = pd.DataFrame.from_dict(mappings)
```
